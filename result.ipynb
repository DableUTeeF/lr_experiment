```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
legends = [l[:-5] for l in os.listdir('logs')]
```

```python
import json
history = []
lookup = {"cos1-02": 0.2, "step-01": 0.1, "cos1-03": 0.2, "step-02": 0.2, "step-03": 0.05, "step-04": 0.01, "step-05": 0.045, "step-06": 0.05}                        
for i in range(len(legends)):
    h = json.load(open(os.path.join('logs', legends[i]+'.json'), 'r'))
    history.append(h)
    legends[i] = legends[i][:-2]+str(lookup[legends[i]])
```

```python
def plot_history(hist, att, axis):
    ylabel = {'oss': 'loss', 'acy': 'accuracy'}
    plt.figure()
    plt.title(att)
    plt.axis(axis)
    for h in hist:
        plt.plot(h[att])
    plt.ylabel(ylabel[att[-3:]])  
    plt.xlabel('epoch')  
    plt.legend(legends)
    plt.show()
```

# Validation Accracy

```python
plot_history(history, 'val_categorical_accuracy', (150, 300, 0.934, 0.942))
```

# Train Accuracy

```python
plot_history(history, 'categorical_accuracy', (-1, 300, 0.99, 1.002))
```

# Validation Loss

```python
plot_history(history, 'val_CrossEntropyLoss', (-1, 300, 0.3, 1))
```

# Train Loss

```python
plot_history(history, 'CrossEntropyLoss', (-1, 300, 1e-9, .05))
```
